```python
for i in categorical_fea:
    val = fea[i].value_counts().sort_values(ascending=False).index[0]
    df_n = fea[i].replace(val1[5],val)
    fea[i] = df_n    
```

```python
y = new_df.loc[:,new_df.columns == 'salary']
x = new_df.loc[:,new_df.columns != 'salary']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
kn = KNeighborsClassifier(n_neighbors=23)
```

```python
kn.fit(X_train,y_train)
```

```python
p = kn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test,p))
```

```python
print(classification_report(y_test,p))
```
